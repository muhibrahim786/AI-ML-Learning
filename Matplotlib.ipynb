```python
#Tight layout is used when the name of any thing in the x and y axis overlaps with other name in the same axis
```

```python
movies=["Money Heistasdfghjklrtr","Young sheldon","Mad unicorn","Game of thrones","Kurulus orhan"]
revenue=[200,400,567,300,800]
plt.plot(movies,revenue)
plt.xlabel("Best Movies")
plt.ylabel("Revenue {in $M}")
plt.title("Best movies and their revenue")
plt.show()
plt.tight_layout()
plt.savefig("final_movies.png")

```
