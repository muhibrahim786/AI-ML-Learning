```python
import seaborn as sns
from sklearn.datasets import make_blobs
X,y=make_blobs(
    n_samples=1000,
    n_features=2,
    centers=4,
    random_state=42
)
```

```python
X
```

```python
y
```

```python
#Visualise the data
sns.scatterplot(x=X[:,0],y=X[:,1])
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(X)
```

```python
#Applying the algorithm using random value of K
from sklearn.cluster import KMeans
K=4
KMeans=KMeans(n_clusters=K,random_state=42)
```

```python
#Label means cluster number
labels=KMeans.fit_predict(X)
```

```python
labels
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],c=labels)
```

#Elbow method

```python
wcss=[]
from sklearn.cluster import KMeans
for k in range(1,21):

  kmeans=KMeans(n_clusters=k)
  kmeans.fit_predict(X)
  wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=list(range(1,21)),y=wcss,marker="o")
```

```python
print(kmeans.inertia_)
```

```python
knee=KneeLocator(range(1,21),wcss,curve="convex",direction="decreasing")
```

```python
print("optimal K = ",knee.elbow)
```

#Silhouette Score

```python
from sklearn.metrics import silhouette_score

```

```python
ss=[]
for k in range(2,21):
  kmeans=KMeans(n_clusters=k)
  labels=kmeans.fit_predict(X)
  score=silhouette_score(X,labels)
  ss.append(score)
```

```python
#plot k and ss
sns.lineplot(x=range(2,21),y=ss,marker="o")
```

#Hierarchical Clustering

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
X=iris.data
y=iris.target
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(X)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,2],c=y)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
#Linear Matrix
Z=linkage(X_scaled,method="ward")
#plot
plt.figure(figsize=(25,10))
dendrogram(Z)
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()


```

```python
#Clustering
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=2)
labels=ac.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,2],c=labels)
```

#DBSCAN Clustering

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(iris.data)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,2])
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.5,min_samples=5)
labels=dbscan.fit_predict(X_scaled)
```

```python
labels
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,2],c=labels)
#
```

```python
#DBSCAN impplementation using non linear data
```

```python
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=300,noise=0.05,random_state=42)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1])
```

```python
X_scaled=sc.fit_transform(X)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,1])
```

```python
#Applying K Means
from sklearn.cluster import KMeans
K=2
kmeans=KMeans(n_clusters=K,random_state=42)
labels=kmeans.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,1],c=labels)
```

```python
#Applying DBSCAN
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.2,min_samples=5)
labels=dbscan.fit_predict(X_scaled)
```

```python
sns.scatterplot(x=X_scaled[:,0],y=X_scaled[:,1],c=labels)
```
