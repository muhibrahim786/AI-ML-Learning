```python
#Scatter Plot(Numerical plots)
tip=sns.load_dataset("tips")


```

```python
#Bivariate Analysis
sns.scatterplot(data=tip,x="total_bill",y="tip")
```

```python
#Multivariate Analysis
sns.scatterplot(data=tip,x="total_bill",y="tip",hue="sex",style="smoker")
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(data=titanic,x="pclass",y="age",kind="bar",hue="sex")
```

```python
sns.catplot(data=titanic,x="pclass",y="fare",kind="bar",hue="who")
```

```python
#Box plot   (Numerical   categorical)
sns.catplot(data=titanic,x="sex",y="age",kind="box",hue="survived")
```

```python
#Distplot     (Numerical    Categorical)
sns.displot(data=titanic,x="age",kind="kde",hue="sex")
```

```python
titanic
```

```python
sns.kdeplot(data=titanic,x="age",hue="sex")
```

```python
#Heat Map (Categorical  categorical)
titanic.head()
```

```python

#Information between two columns
pd.crosstab(titanic["pclass"],titanic["survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["pclass"],titanic["survived"]))
```

```python
(titanic.groupby("pclass").mean()["survived"]*100).plot(kind="bar")
```

```python
#Cluster Map (Categorical  categorical)
pd.crosstab(titanic["pclass"],titanic["survived"])
```

```python
pd.crosstab(titanic["sibsp"],titanic["survived"])
```

```python
#Cluster Map (Categorical  categorical)
sns.clustermap(pd.crosstab(titanic["sibsp"],titanic["survived"]))
```

```python
#Pair plot
sns.pairplot(titanic)
```

```python
iris=sns.load_dataset("iris")
sns.pairplot(iris,hue="species")
```
